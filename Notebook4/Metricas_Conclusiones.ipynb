```python
import pandas as pd
from jinja2 import Template
# Cargar las métricas desde el archivo CSV a un DataFrame
metricas_df = pd.read_csv("df_metricas.csv")
metricas_df
```

```python
# Funcion para decargar html con las métricas
html_content = """
<!DOCTYPE html>
<html>
<head>
    <title>Resultados</title>
    <style>
        body {
            font-family: Arial, sans-serif;
        }
        h3 {
            color: black;
            text-align: center;
        }
        table {
            border-collapse: collapse;
            width: 60%; /* Reducir el ancho de la tabla */
            margin: 20px auto;
            border: 1px solid #ddd;
            max-height: 400px; /* Establecer una altura máxima para la tabla */
            overflow-y: auto; /* Agregar barra de desplazamiento vertical si el contenido excede la altura */
        }
        th, td {
            padding: 4px; /* Reducir el padding para hacer las filas más estrechas */
            font-size: 12px; /* Reducir el tamaño de la fuente */
            text-align: left;
            border-bottom: 1px solid #ddd;
        }

    </style>
</head>
<body>

    <h3>Resultados globales</h3>
    {{ results_globales|safe }}

    <h3>Resultados locales</h3>
    {{ results_por_label|safe }}

    <h3>Resultados cm</h3>
    {{ results_cm|safe }}

</body>
</html>
"""
# Crear la plantilla Jinja2
template = Template(html_content)


# Obtener las etiquetas de las métricas por etiqueta
etiquetas = ["label 0/1"]  # Reemplaza esto con las etiquetas reales

# Crear un DataFrame para las etiquetas
etiquetas_df = pd.DataFrame({'Etiqueta': etiquetas})

# Concatenar el DataFrame de etiquetas con las métricas por etiqueta
etiquetas_metrics_df = pd.concat([etiquetas_df, metricas_df[["eval_precision_por_label", "eval_recall_por_label", "eval_f1_por_label"]]], axis=1)

# Datos a utilizar en el reporte
data = {
    'results_globales': metricas_df[["eval_accuracy", "eval_precision_global", "eval_recall_global", "eval_f1_global"]].to_html(index=False, escape=False),
    'results_por_label': etiquetas_metrics_df[["Etiqueta", "eval_precision_por_label", "eval_recall_por_label", "eval_f1_por_label"]].to_html(index=False, escape=False),
    'results_cm': metricas_df[["eval_matriz_confusion"]].to_html(index=False, escape=False, header=False)
}

# Renderizar el reporte HTML con los datos
html_output = template.render(data)

# Guardar el reporte en un archivo HTML
with open('Métricas.html', 'w') as file:
    file.write(html_output)
```
